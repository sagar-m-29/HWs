```python
class Board():
    def __init__(self):
        
        self.values = [" "]*10

    def display(self):
        self.board = f"""
        {self.values[1]}|{self.values[2]}|{self.values[3]}
        _ _ _
        {self.values[4]}|{self.values[5]}|{self.values[6]}
        _ _ _
        {self.values[7]}|{self.values[8]}|{self.values[9]}
        """

        print(self.board)
        
    def place(self, position, character):
        self.values[position] = character

    def game_is_won(self):
        return (self.values[1] == self.values[2] == self.values[3] != " " or
            self.values[4] == self.values[5] == self.values[6] != " " or
            self.values[7] == self.values[8] == self.values[9] != " " or
            self.values[1] == self.values[4] == self.values[7] != " " or
            self.values[2] == self.values[5] == self.values[8] != " " or
            self.values[3] == self.values[6] == self.values[9] != " " or
            self.values[1] == self.values[5] == self.values[9] != " " or
            self.values[3] == self.values[5] == self.values[7] != " ")
```

```python
class Player():
    def __init__(self,name,symbol):
        self.name = name
        self.symbol = symbol

    def player_name(self):
        return self.name
    
    def player_symbol(self):
        return self.symbol
    
    
```

```python
class TicTacToe():
    def __init__(self):
        self.board = Board()


    def player_input(player_number):
        p_name = input(f"{player_number}, what's your name?") 
        while True:
            p_symbol = input(f"{p_name}, what's your symbol? x or o?") 
            if p_symbol == "X" or p_symbol == "x" or p_symbol ==  "o" or p_symbol == "O":
                break
            else:
                print("Not a valid symbol, please try again.")

        return Player(p_name,p_symbol)
    
    def player_input_v2(self,player):

        while True:
            position = int(input(f"{player.player_name()}, what is your position?"))

            if self.board.values[position] != " ":
                print("Error, please try again")
            else:
                self.board.place(position, player.player_symbol())
                break


    def play(self):

        self.player_1 = TicTacToe.player_input("Player 1")
        self.player_2 = TicTacToe.player_input("Player 2")

        self.board.display()

        turn = 1

        while True:
            if turn == 1:
                self.player_input_v2(self.player_1)

                turn = 0
            else:
                self.player_input_v2(self.player_2)

                turn = 1

            self.board.display()

            if self.board.game_is_won():
                
                if turn == 1:
                    winner = self.player_2
                else:
                    winner = self.player_1

                print(f"Game has ended. Winner is {winner.player_name()}!")

                break

```
