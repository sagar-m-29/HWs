```python
class Board():
    def __init__(self):
        
        self.values = [" "]*9

        self.board = f"""
         | |
        _ _ _
         | |
        _ _ _
         | |
        """

    def update_board(self):

        self.board = f"""
        {self.values[0]}|{self.values[1]}|{self.values[2]}
        _ _ _
        {self.values[3]}|{self.values[4]}|{self.values[5]}
        _ _ _
        {self.values[6]}|{self.values[7]}|{self.values[8]}
        """

    def display(self):
        print(self.board)
        
    def place(self, position, character):
        self.values[position] = character
        self.update_board()
    
```
