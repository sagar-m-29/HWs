```python
class Board():
    def __init__(self):
        
        self.values = [" "]*10

    def display(self):
        self.board = f"""
        {self.values[1]}|{self.values[2]}|{self.values[3]}
        _ _ _
        {self.values[4]}|{self.values[5]}|{self.values[6]}
        _ _ _
        {self.values[7]}|{self.values[8]}|{self.values[9]}
        """

        print(self.board)
        
    def place(self, position, character):
        self.values[position] = character
        self.display()
    
```

```python
class Player():
    def __init__(self,name,symbol):
        self.name = name
        self.symbol = symbol

    def player_name(self):
        return self.name
    
    def player_symbol(self):
        return self.symbol
```

```python
class TicTacToe():

    def __init__(self):
        self.Board = Board()

    def player_name_symbol(self):
            p1_name = input("p1, whats name?")
            p2_name = input("p2, whats name?")

            p1_symbol = input(f"{p1_name}, whats symbol?")
            p2_symbol = input(f"{p2_name}, whats symbol?")

            self.p1 = Player(p1_name,p1_symbol)
            self.p2 = Player(p2_name,p2_symbol)

# past_player_inputs array never updates, tried many locations and for else breaking loop and doing outside as well
    def player_input(self, player):

        past_player_inputs = []
        condition = 1

        while condition == 1:

            player_input = int(input(f"{player.player_name()}, where do you want to place?"))

            if player_input > 9 or player_input < 1:
                print("Error, choice of placement not in board")
                player_input = int(input(f"{player.player_name()}, where do you want to place?"))

            elif player_input in past_player_inputs:
                print("Error, that place is occupied, please place elsewhere")
                player_input = int(input(f"{player.player_name()}, where do you want to place?"))

            else:
                self.Board.place(player_input, player.player_symbol())
                print(past_player_inputs)
                past_player_inputs.append(player_input)
                condition = 0

    def play(self):

        turn = 1

        self.Board.display()

        while True:
            if turn == 1:
                self.player_input(self.p1)

                turn = 0

            else:
                self.player_input(self.p2)
                
                turn = 1

            if (self.Board.values[1] == self.Board.values[2] == self.Board.values[3] != " " or
            self.Board.values[4] == self.Board.values[5] == self.Board.values[6] != " " or
            self.Board.values[7] == self.Board.values[8] == self.Board.values[9] != " " or
            self.Board.values[1] == self.Board.values[4] == self.Board.values[7] != " " or
            self.Board.values[2] == self.Board.values[5] == self.Board.values[8] != " " or
            self.Board.values[3] == self.Board.values[6] == self.Board.values[9] != " " or
            self.Board.values[1] == self.Board.values[5] == self.Board.values[9] != " " or
            self.Board.values[3] == self.Board.values[5] == self.Board.values[7] != " "):
                print("game had ended")
                break
```

```python
t = TicTacToe()

t.player_name_symbol()

t.play()
```
