```python
import pandas as pd

fpl_data = pd.read_csv("cleaned_players.csv")

fpl_data["full_name"] = fpl_data["first_name"] + " " + fpl_data["second_name"]

fpl_data.set_index("full_name", inplace=True)

fpl_data.head()
```

```python
def get_player():
    while True:
        player_name = input("Player Name: ")
        if player_name in fpl_data.index.values:
            print(fpl_data.loc[player_name])
            break
        elif player_name in fpl_data["first_name"].values: 
            print(fpl_data[fpl_data["first_name"] == player_name]) 
            break
        elif player_name in fpl_data["second_name"].values:
            print(fpl_data[fpl_data["second_name"] == player_name])
            break
        else:
            print("This name doesn't exist in the dataset, try again.")

get_player()
```

```python
def top_player_on_a_stat():
    while True:
        stat = input("What is the stat?: ")
        if stat in fpl_data.columns:
            sorted_data = fpl_data.sort_values(by=stat, ascending=False)
            top_player = sorted_data.iloc[0]
            return f"The player with the top {stat} is {top_player['first_name']} {top_player['second_name']} with {top_player[stat]} {stat}."
        else:
            print("This stat doesn't exist, try again")

print(top_player_on_a_stat())
```

```python
def best_player_per_position():
    while True:
        position = input("What is the players' position? ")
        budget = float(input("What is your budget? "))
        if position in fpl_data["element_type"].values:
        # sorted_data = fpl_data[fpl_data["element_type"] == position]
        # sorted_data = fpl_data[fpl_data["now_cost"] <= budget]
            sorted_data = fpl_data[(fpl_data["element_type"] == position) & (fpl_data["now_cost"] <= budget)]
            filtered = sorted_data.sort_values(by="now_cost", ascending=False)
            return filtered.iloc[0]
        else:
            print("Position wasn't correctly identified.")
        
        return filtered.iloc[0]

best_player_per_position()

```
