```python
import pandas as pd
```

```python
class Player:
    def __init__(self, first_name, second_name, data):
        self.first_name = first_name
        self.second_name = second_name
        self.data = data

    def full_name(self):
        return f"{self.first_name}" + " " + f"{self.second_name}"

    def __str__(self):
        return self.data
```

```python
class Players:
    def __init__(self, file_path):
        self.fpl_data = pd.read_csv(file_path)
        self.fpl_data["full_name"] = self.fpl_data["first_name"] + " " + self.fpl_data["second_name"]
        self.fpl_data.set_index("full_name", inplace=True)

    def get_player(self, name):
        player_data_list = []
        other_list = []
        if name in self.fpl_data.index:
            player_data_list.append(self.fpl_data.loc[name])
        if name in self.fpl_data["first_name"].values:
            player_data = self.fpl_data[self.fpl_data["first_name"] == name]
            other_list.extend(player_data["first_name"] + " " + player_data["second_name"])
        if name in self.fpl_data["second_name"].values:
            player_data = self.fpl_data[self.fpl_data["second_name"] == name]
            other_list.extend(player_data["first_name"] + " " + player_data["second_name"])
        for data in player_data_list:
            print(data)
        for player_name in other_list:
            while None in other_list:
                other_list.remove(None)
            print(self.fpl_data.loc[player_name])

    def top_player_on_stat(self, stat):
        if stat in self.fpl_data.columns:
            sorted_data = self.fpl_data.sort_values(by=stat, ascending=False)
            top_player_data = sorted_data.iloc[0]
            return Player(top_player_data['first_name'], top_player_data['second_name'], top_player_data)
        else:
            print("This stat doesn't exist, try again.")

    def best_player_per_position(self, position, budget):
        if position in self.fpl_data["element_type"].values:
            sorted_data = self.fpl_data[(self.fpl_data["element_type"] == position) & (self.fpl_data["now_cost"] <= budget)]
            top_player_data = sorted_data.sort_values(by="now_cost", ascending=False).iloc[0]
            return Player(top_player_data['first_name'], top_player_data['second_name'], top_player_data)
        else:
            print("No players found with this position and/or budget.")

    def compare_players(self, player1_name, player2_name):
        if player1_name in self.fpl_data.index and player2_name in self.fpl_data.index:
            player1_data = self.fpl_data.loc[player1_name]
            player2_data = self.fpl_data.loc[player2_name]
            player1 = Player(player1_data['first_name'], player1_data['second_name'], player1_data)
            player2 = Player(player2_data['first_name'], player2_data['second_name'], player2_data)
            print(player1)
            print(player2)
        else:
            full_names = []

            if player1_name in self.fpl_data["first_name"].values:
                player1_first_name = self.fpl_data[self.fpl_data["first_name"] == player1_name]
                full_names.extend(player1_first_name.index.tolist())

            if player1_name in self.fpl_data["second_name"].values:
                player1_second_name = self.fpl_data[self.fpl_data["second_name"] == player1_name]
                full_names.extend(player1_second_name.index.tolist())

            if player2_name in self.fpl_data["first_name"].values:
                player2_first_name = self.fpl_data[self.fpl_data["first_name"] == player2_name]
                full_names.extend(player2_first_name.index.tolist())

            if player2_name in self.fpl_data["second_name"].values:
                player2_second_name = self.fpl_data[self.fpl_data["second_name"] == player2_name]
                full_names.extend(player2_second_name.index.tolist())

            if len(full_names) > 0:
                comparison_data = self.fpl_data.loc[full_names]
                print(comparison_data)
            else:
                print("No valid players found in dataset.")

    def run(self):
        while True:
            choice = input("Choose an option: 1 for Get Player // 2 for Top Player on Stat // 3 for Best Player per Position // 4 for Compare Players // 5 for Exit: ")
            if choice == '1':
                player_name = input("Player Name: ")
                print(self.get_player(player_name))
            elif choice == '2':
                stat = input("What is the stat?: ")
                player = self.top_player_on_stat(stat)
                print(f"The player with the top {stat} is {player.full_name()} with {player.data[stat]} {stat}.")
            elif choice == '3':
                position = input("What is the players' position? ")
                budget = float(input("What is your budget? "))
                print(self.best_player_per_position(position, budget))
            elif choice == '4':
                player1 = input("Enter 1st player: ")
                player2 = input("Enter 2nd player: ")
                self.compare_players(player1, player2)
            elif choice == '5':
                break
            else:
                print("Invalid choice, please try again.")
```

```python
players = Players("cleaned_players.csv")
players.run()
```
